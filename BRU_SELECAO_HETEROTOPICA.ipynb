## Funções

```python
def cleanup_grq(df, df_subset):
    
    df.loc[df_subset,'FGRA'] = np.nan
    for fx in faixa:       
        df.loc[df_subset,'FQ' + fx] = np.nan
        df.loc[df_subset,'G' + fx] = np.nan
        for v in var6_quim + var3_quim:  
            df.loc[df_subset, 'A' + v + fx] = np.nan
            df.loc[df_subset, v + fx] = np.nan
    return df
```

## Finally combine loop from 10 to 100% sample and save to csv...

```python
cac=pd.read_csv('brubru_het_cac_entry_filtro_HETERO.csv', low_memory=False, na_values='-99', encoding="latin1")
cac.rename(columns=str.upper, inplace=True)
```

```python
bs=pd.DataFrame()
for x in range(10,100,10):
  print(str(x) + '% REGISTERS SAMPLED')
  s = cac.sample(frac=x/100, random_state=2609)

  s['TAG']=x 
  tx = s.describe()
  tx['TAG'] =  x
  tx['NEWINDEX'] =  tx.index + '_' + str(x)
  tx.set_index('NEWINDEX', inplace=True)
  bs=bs.append(tx)  
  smp = s.loc[:, 'TAG']

  joined= cac.join(smp)

  i = (joined['TAG'].notna())
  smp2 = cleanup_grq(joined, i)
  smp2.to_csv("test_BRU_sampled_df_"+str(x)+".csv") 
  
  if(x==90):
    bs.to_csv('sampling_stats.csv')

else:
  print("Finally finished!")
```
