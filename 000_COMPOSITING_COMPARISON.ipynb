```python
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
```

```python
chart_size=[10, 10]
```

```python
path = "C:\DADOS\ANDERSON\PROJ_VULCAN\SN3_1121\Arq_Vulcan\\"
```

```python
fso = pd.read_csv(path + "n3flp_cac_entry.csv", low_memory=False, na_values='-99', encoding="latin1")
fso.rename(columns=str.upper, inplace=True)
```

```python
cac = pd.read_csv(path + "sn3flp_fso_assay.csv", low_memory=False, na_values='-99', encoding="latin1")
cac.rename(columns=str.upper, inplace=True)
```

```python
from ipywidgets import interactive

def bin_widget(bw=0.1):
    sea.histplot(data=fso["FEGL"],binwidth=bw,kde=True, color='r')
```

```python
interactive(bin_widget, bw=(0, 1, 0.05))
```

```python
fig, ax = plt.subplots(figsize=chart_size)
sea.histplot(data=fso["FEGL"],binwidth=1,kde=True, color='r')
sea.histplot(data=cac["FEGL"],binwidth=1,kde=True, color='g')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('SAMPLE COUNT')
ax.set_title('OXIDE BEARING ROCKS FEGL')

plt.show()
fig.savefig('TEST_HISTOGRAM_FEGL.jpeg', transparent=None)
```
