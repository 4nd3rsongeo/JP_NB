```python
import pandas as pd

# CLASSIFIED DATA
arquivo = pd.read_csv('C:/DADOS/ML/1_Historical_Geochemistry_Study.csv')
# LOAD UNCLASSIFIED DATA
arquivo2 = pd.read_csv('C:/DADOS/ML/2_Recent_whole_rock_sampling_from_drilling.csv')
```

```python
arquivo.head()
```

```python
# ATRIBUTE NUMBERS TO CATEGORICAL VARIABLE

#arquivo['Chem_strat'] =  arquivo['Chem_strat'].replace('Punk Sediments',0)
#arquivo['Chem_strat'] =  arquivo['Chem_strat'].replace('Lenin Basalt',1)

# THERE MUST BE ANOTHER WAY TO DO THIS... ASKING GOOGLE ...
arquivo.Chem_strat = pd.Categorical(arquivo.Chem_strat)

arquivo['codes'] = arquivo.Chem_strat.cat.codes
```

```python
arquivo.head()
```

```python
# SPLITING VARIABLES AND TARGET
y = arquivo['codes']
x = arquivo.drop(['codes','Chem_strat'], axis =1)

x.head()
```

```python
y.head()
```

```python
# LOADING SKLEARN... 70/30%

from sklearn.model_selection import train_test_split

# criando conjuntos de dados de treino e teste com 70% dos dados para treino
# e 30% de dados para teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)

# MAKING SURE DATASET WAS SPLITED AS EXPECTED...
arquivo.shape
```

```python
x_teste.shape

```

```python
y_teste.shape
```

```python
arquivo['codes'].max()
```

```python
x_teste.head()
```

```python
y_teste.head()
```

```python
x_treino.shape
```

```python
# Classificação por árvores de decisão
from sklearn.ensemble import RandomForestClassifier
# Criação do modelo:
modelo = RandomForestClassifier()
modelo.fit(x_treino, y_treino)

# Imrpimindo Resultados:
resultado = modelo.score(x_teste, y_teste)
print("Acurácia: ", resultado)
```
